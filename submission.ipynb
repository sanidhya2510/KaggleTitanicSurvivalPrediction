```python
import pandas as pd 
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
```

```python
trainData = pd.read_csv('train.csv')
testData = pd.read_csv('test.csv')
trainData.head()
```

```python
trainResult = trainData['Survived']
print(testData)
```

```python
newTrainData = trainData[['PassengerId' , 'Pclass', 'Sex', 'Age' ,'SibSp' , 'Parch' , 'Fare' , 'Embarked']]
newTestData = testData[['PassengerId' , 'Pclass', 'Sex', 'Age' ,'SibSp' , 'Parch' , 'Fare' , 'Embarked']]
newTrainData
```

```python
x = (newTrainData.dtypes == 'object')
objCols = list(x[x].index)
objCols
```

```python
ordinalEncoder = OrdinalEncoder()

newTrainData[objCols] = ordinalEncoder.fit_transform(newTrainData[objCols])
newTestData[objCols] = ordinalEncoder.transform(newTestData[objCols])

print(newTrainData)
```

```python
colsMissing = [col for col in newTrainData.columns
               if newTrainData[col].isnull().any()]
print(colsMissing)
imputer = SimpleImputer()
newImputedData = pd.DataFrame(imputer.fit_transform(newTrainData))
newImputerTestData = pd.DataFrame(imputer.transform(newTestData))

newImputedData.columns = newTrainData.columns
newImputerTestData.columns = newTestData.columns
```

```python
model = RandomForestClassifier(random_state=0, n_estimators=100)
model.fit(newImputedData, trainResult)
result = model.predict(newImputerTestData)
print(result)
print(trainResult)

output = pd.DataFrame({'PassengerId' : newImputerTestData.PassengerId.astype(int), 'Survived' : result.astype(int)})
output.to_csv('submission.csv', index = False)
print("Your submission was successfully saved!")
```
